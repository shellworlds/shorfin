```
# Install required dependencies
!pip install -q numpy matplotlib networkx

# Import and run specific cells from the notebook
import json
import nbformat

# Load the notebook
with open('shor_algorithm_simulation.ipynb', 'r') as f:
    nb = nbformat.read(f, as_version=4)

# Find and execute specific code cells
print(f"Loaded notebook: {len(nb.cells)} cells")

# Execute initialization cells
execution_count = 0
for i, cell in enumerate(nb.cells[:10]):  # First 10 cells
    if cell.cell_type == 'code':
        try:
            exec(cell.source)
            execution_count += 1
        except Exception as e:
            print(f"Cell {i} error: {e}")

print(f"Successfully executed {execution_count} code cells")
```
